### Identify Clusters (via KMeans)

```python
# Derive List of Coordinates
y = bldngs['latitude'].tolist()
x = bldngs['longitude'].tolist()
x_y = zip(x, y)
```

```python
# Choose K
k_optimal = 5
```

#### Graph Clusters (k = k_optimal)

```python
kmeans = KMeans(n_clusters=k_optimal, random_state=0)
kmeans.fit(x_y)
y_kmeans = kmeans.predict(x_y)
```

```python
# color-coded
centers = kmeans.cluster_centers_
if graphs_on:
    plt.scatter(x, y, c=y_kmeans, s=0.000001, cmap='viridis') 
    # monochrome
    # plt.scatter(x, y, s=0.0001, c="blue")
    # plt.scatter(centers[:, 0], centers[:, 1], c='orange', s=150)
    plt.scatter(centers[:, 0], centers[:, 1], c='black', s=100, alpha=.5)
```

```python
# Export Cluster Assignments for Reference Later On
clusters = pd.DataFrame({'cluster_k' + str(k_optimal):y_kmeans\
                   , 'longitude':x, 'latitude':y })
clusters.to_csv(data_folder + 'regions/cluster_assignments.csv')
```

### Construct Polygons

```python
clusters = np.array(zip(centers[:, 0], centers[:, 1]))
vor = Voronoi(clusters)
```

```python
# plot
regions, vertices = voronoi_finite_polygons_2d(vor)

# Find Polygons
polygons = []
for region in regions:
    polygon = vertices[region]
    polygons.append(polygon)
    if graphs_on:
        plt.fill(*zip(*polygon), alpha=0.4)

if graphs_on:
    plt.plot(clusters[:,0], clusters[:,1], 'ko')
    plt.xlim(min(x), max(x))
    plt.ylim(min(y),max(y))

    plt.show()
```

```python
i = 0
polygons_json = []
for polygon in polygons:
    i+=1
    polygon_shapely = mapping(Polygon(polygon))
    polygon_json = json.dumps(polygon_shapely)
#     polygons_json.append(polygon_json)
#     polygons_str = ''.join(polygons_json)
    with open(data_folder + 'regions/region_' + str(i)\
              + '.geojson', 'w') as output_file:
        output_file.write(polygon_json) 
```

```python
regions = gpd.read_file(data_folder + 'regions/region_1.geojson')
for region in range(2, k_optimal + 1):
    regions = regions.append(gpd.read_file(data_folder + 'regions/region_'\
                                           + str(region) + '.geojson'))
# regions.plot()
# plt.xlim(min(x), max(x))
# plt.ylim(min(y),max(y))
try:
    regions.to_file(data_folder + 'regions/regions.geojson', driver='GeoJSON')
except:
    print """
                'regions/regions.geojson' Geofile already exists. 
                Remove old version for replacement.
            """
cluster_polygons = gpd.read_file(data_folder + 'regions/regions.geojson')
```

```python
states_borders = gpd.read_file(data_folder\
                              +'states_outlines/cb_2017_us_state_5m.shp')
i = 0
for state, state_abbv in states:
    state_outline = states_borders[states_borders.NAME==state]
    if i == 0:
        state_group = state_outline
    else:
        state_group = gpd.overlay(state_group, state_outline\
                                , how='union')
    i += 1

intersection = gpd.overlay(state_group, cluster_polygons\
                           , how='intersection')
if graphs_on:
    intersection.plot()
```
